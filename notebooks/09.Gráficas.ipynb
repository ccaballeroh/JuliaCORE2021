# Graficar

## Lo básico
Hay algunas formas de graficar en Julia (incluyendo el llamar a PyPlot).<br>

Aquí veremos cómo usar `Plots.jl`. Si aún no está instalado, necesitas usar el gestor de paquetes para instalarlo y Julia lo precompilará por ti la primera vez ue lo uses:
Here we'll show you how to use `Plots.jl`.  If it's not installed yet, you need to use the package manager to install it, and Julia will precompile it for you the first time you use it:

````julia
using Pkg
Pkg.add("Plots")
using Plots
````

Una de las ventajas de `Plots.jl` es que permite intercambiar de *backend* de forma transparente. En este *notebook* probaremos los *backends* `gr()` y el `plotlyjs()`.

En el nombre de la ciencia, usemos este *notebook* para examinar la relación entre la temperatura global y el número de piratas entre los años 1860 y 2000.

````julia
globaltemperatures = [14.4, 14.5, 14.8, 15.2, 15.5, 15.8]
numpirates = [45000, 20000, 15000, 5000, 400, 17];
````

`Plots` soporta varios *backends*, es decir, bibliotecas que se encargan del dibujo, todos con la misma API. Para comenzar, intentemos el *backend* GR. Esto se hace con una llamada a `gr()`:

````julia
#gr()
#using Pkg
Pkg.add("PlotlyJS")
plotlyjs()
````

Y ahora usamos los comandos como `plot` y `scatter` para generar gráficas.

````julia
plot(numpirates, globaltemperatures, label="line")
scatter!(numpirates, globaltemperatures, label="points")
````

Recordemos que el `!` al final del nombre de las funciones nos indica que la función modifica el dato que se le pasa como argumento. Así que los puntos del *scatter* de van a agregar a la gráfica que ya existía.

En contraste, observa qué sucede cuando se reemplaza `scatter!` con la función `scatter`.

A continuación, actualicemos la gráfica con `xlabel!`, `ylabel!` y `title!`.

````julia
xlabel!("Number of Pirates [Approximate]")
ylabel!("Global Temperature (C)")
title!("Influence of pirate population on global warming")
````

Esto no se ve bien aún. El número de piratas ha disminuido desde 1860, así que leer la gráfica de izquierda a derecha es como ver hacia atrás en el tiempo en lugar de hacia adelante. Volteemos el eje equis para ver mejor cómo las poblaciones de piratas han *causado* que las temperaturas globales cambien en el tiempo.

````julia
xflip!()
````
